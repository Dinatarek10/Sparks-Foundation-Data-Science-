```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import metrics
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
data
```

```python
X =data['Hours']
Y = data['Scores']
```

**Data Visualization**

```python
plt.scatter(X, Y)
plt.xlabel('Hours', fontsize = 20)
plt.ylabel('Scores', fontsize = 20)
plt.show()
```

**Data preparation**

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
print(X)
print(Y)
```

**Split and Training**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
regressionModel = linear_model.LinearRegression()
regressionModel.fit(X_train,y_train)
```

```python
plt.scatter(X, Y)
plt.xlabel('Hours', fontsize = 20)
plt.ylabel('Scores', fontsize = 20)
plt.plot(X,regressionModel.coef_*X+regressionModel.intercept_)
plt.show()
```

**Making Predictions**

```python
predictions = regressionModel.predict(X_test)
```

```python
df = pd.DataFrame({'Actual':y_test, 'Predicted':predictions})
df
```

**Model Evaluation**

```python
print('Mean Absolute Error', metrics.mean_absolute_error(y_test, predictions))
```

**Predict the Score if a student studied 9.25 hours**

```python
pred = regressionModel.predict([[9.25]])
pred[0]
```
